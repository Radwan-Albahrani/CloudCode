# All our necessary Imports

```python
from skimage.io import imread
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
from skimage.morphology import erosion, disk, dilation, opening, closing
```

# A function to plot 2 images as a comparison

```python
def plot_comparison(original, filtered, filter_name):

    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4), sharex=True,
                                   sharey=True)
    ax1.imshow(original, cmap=plt.cm.gray)
    ax1.set_title('original')
    ax1.axis('off')
    ax2.imshow(filtered, cmap=plt.cm.gray)
    ax2.set_title(filter_name)
    ax2.axis('off')
```

# Applying Erosion followed by Dilation with different sized to achieve desired result

```python
im = rgb2gray(imread('./lab_09_images/Image_1.jpg'))
imBW = im

im1 = erosion(imBW, disk(7))
im2 = dilation(im1, disk(14))

plot_comparison(imBW, im2, "Erosion + Dilation: Opening")
```

# Applying opening process (Same disk size)

```python
im = rgb2gray(imread('./lab_09_images/Image_1.jpg'))
imBW = im

im1 = opening(imBW, disk(8))

plot_comparison(imBW, im1, "Opening")
```

# Applying Dilation followed by erosion with different sizes to achieve desired result

```python
im = rgb2gray(imread('./lab_09_images/Image_1.jpg'))
imBW = im

im1 = dilation(imBW, disk(18))
im2 = erosion(im1, disk(7))

plot_comparison(imBW, im2, "Dilation + Erosion: Closing")
```

# Applying Closing Process (Same Size)

```python
im = rgb2gray(imread('./lab_09_images/Image_1.jpg'))
imBW = im

im1 = closing(imBW, disk(16))

plot_comparison(imBW, im1, "Closing")
```
