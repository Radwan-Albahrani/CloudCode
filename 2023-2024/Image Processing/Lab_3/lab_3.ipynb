```python
# Take the image
import cv2 as cv
import sys
from matplotlib import pyplot as plt

try:
    img = cv.imread("mario.png")

except IOError:
    print("Unable to load image")
    sys.exit(1)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
rotatedImage = cv.rotate(img, cv.ROTATE_180)
plt.imshow(rotatedImage)
```

```python
cropped_image = img[0:150, 50:200]
plt.imshow(cropped_image)
plt.show()

rotateUpsideDown = cv.rotate(cropped_image, cv.ROTATE_180)
plt.imshow(rotateUpsideDown)
plt.show
```

```python
grayImage = cv.cvtColor(cropped_image, cv.COLOR_RGB2GRAY)
plt.imshow(grayImage, cmap="gray")
plt.show()
```

```python
scale_percent = 220 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)

# resize image
resized = cv.resize(img, dim, interpolation = cv.INTER_AREA)

plt.imshow(resized)
plt.show()
```
