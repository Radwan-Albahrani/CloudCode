# Task 1

```python
# Load the data
import pandas as pd
df = pd.read_csv('data/IMDB Dataset.csv', encoding='latin1')
reviews = df['review'].str.cat(sep='\n')
print(reviews[:2000])
```

```python
import re
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

cleaned_reviews = reviews
cleaned_reviews = re.sub(r'<br />', ' ', cleaned_reviews)  # Replace <br /> with a space
cleaned_reviews = re.sub(r'[^a-zA-Z\s]', '', cleaned_reviews)  # Remove non-alphabetic characters
cleaned_reviews = cleaned_reviews.lower()  # Convert to lowercase

# Remove stopwords
cleaned_reviews = ' '.join([word for word in cleaned_reviews.split() if word not in stop_words])
```

```python
print(cleaned_reviews[:2000])
```

```python
total_reviews = len(df)
average_word_per_review = len(cleaned_reviews.split()) / total_reviews
positive_reviews = df['review'].str.contains('good').sum()
negative_reviews = df['review'].str.contains('bad').sum()
print(f'Total reviews: {total_reviews}')
print(f'Average words per review: {average_word_per_review:.2f}')
print(f'Positive reviews: {positive_reviews}')
print(f'Negative reviews: {negative_reviews}')
```

```python
from collections import Counter
import matplotlib.pyplot as plt

word_freq = Counter(cleaned_reviews.split())
word_freq = word_freq.most_common(10)
words = [word[0] for word in word_freq]
freq = [word[1] for word in word_freq]

plt.bar(words, freq)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Word Frequency')
```

# Task 2

```python
df = pd.read_csv('data/poems/Poem_classification - train_data.csv')
poems = df['Poem'].str.cat(sep='\n')
```

```python
cleaned_poems = poems
cleaned_poems = re.sub(r'<br />', ' ', cleaned_poems)  # Replace <br /> with a space
cleaned_poems = re.sub(r'[^a-zA-Z\s]', '', cleaned_poems)  # Remove non-alphabetic characters
cleaned_poems = cleaned_poems.lower()  # Convert to lowercase

# Remove stopwords
cleaned_poems = ' '.join([word for word in cleaned_poems.split() if word not in stop_words])
```

```python
print(cleaned_poems[:2000])
```

```python
import nltk

tokens = nltk.word_tokenize(cleaned_poems)
bigrams = list(nltk.bigrams(tokens))
bigram_freq = Counter(bigrams)
bigram_freq = bigram_freq.most_common(10)
bigrams = [' '.join(bigram) for bigram, freq in bigram_freq]
freq = [freq for bigram, freq in bigram_freq]
# Make the plot as wide as possible
plt.figure(figsize=(15, 5))
plt.bar(bigrams, freq)
plt.xlabel('Bigrams')
plt.ylabel('Frequency')
plt.title('Bigram Frequency')
```
